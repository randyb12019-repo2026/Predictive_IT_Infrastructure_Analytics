### **4.1. Cargar predicciones del modelo seleccionado (en Pickle)**

En primer lugar, se muestra la representación del objeto **RandomForestClassifier** cargado desde el archivo Pickle, incluyendo todos sus parámetros de configuración:

- **`n_estimators=200`**: el modelo está compuesto por 200 árboles de decisión.
- **`class_weight='balanced_subsample'`**: ajusta los pesos de forma automática para compensar el desbalance de clases.
- **`random_state=42`**: semilla fija que garantiza la reproducibilidad de los resultados.
- **`max_depth`, `min_samples_split`, `min_samples_leaf`**: parámetros con valores por defecto que controlan la profundidad y poda de los árboles.

En segundo lugar, el mensaje **"Modelo cargado correctamente."** confirma que el proceso de deserialización del archivo Pickle se completó sin errores.

La correcta carga del modelo validado permite utilizarlo inmediatamente para realizar predicciones sobre nuevos datos de infraestructura IT, sin necesidad de re-entrenarlo. Esto reduce el tiempo de respuesta y garantiza la consistencia entre los resultados del notebook de modelado y los que se mostrarán en el dashboard.
