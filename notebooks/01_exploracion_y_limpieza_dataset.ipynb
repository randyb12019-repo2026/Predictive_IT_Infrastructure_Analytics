### **12. Correlaciones**

```python
# Calcular la matriz de correlación entre las variables numéricas
corr = numericas.corr()

# Definir el tamaño de la figura donde se mostrará el mapa de calor
plt.figure(figsize=(8,6))

# Generar un mapa de calor para visualizar la intensidad de las correlaciones
sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=False
)

# Añadir un título descriptivo al gráfico
plt.title("Matriz de correlación")

# Mostrar el mapa de calor generado
plt.show()
```

```python
# Muestra exactamente qué variables tienen más relación con status
corr["status"].sort_values(ascending=False)
```

#### **12.1. Análisis de las variables tienen más relación con status**

El análisis indica que las métricas más relevantes para explicar el estado operativo de la infraestructura son:

- Temperatura del sistema (`temperature`) -> Correlación: 0.117

Aunque las correlaciones observadas no son especialmente elevadas, estas variables constituyen los mejores candidatos iniciales para la construcción de modelos predictivos orientados a la detección temprana de estados anómalos.

```python
# Muestra el heatmap de correlación con valores para una mejor interpretación visual
corr = numericas.corr()

# Definir el tamaño de la figura donde se mostrará el mapa de calor
plt.figure(figsize=(8,6))

# Generar un mapa de calor para visualizar la intensidad de las correlaciones con anotaciones
sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=True,
    fmt=".2f"
)

# Añadir un título descriptivo al gráfico
plt.title("Matriz de correlación")

# Mostrar el mapa de calor generado
plt.show()
```
