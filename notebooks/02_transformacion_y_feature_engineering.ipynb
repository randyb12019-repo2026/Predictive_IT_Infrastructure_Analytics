### **Fase 4. Transformación de Datos**
### **1. Importación de librerias:**

```python
# Importación de librerias
import pandas as pd # Librería para trabajar con datos en forma de tablas
import numpy as np # Librería para operaciones numéricas
import os # Permite trabajar con rutas y carpetas del sistema
```

### **2. Carga del DataSet:**

```python
# Carga del dataset
df = pd.read_csv("../data/raw/Big_data_dataset.csv")

# Mostrar las primeras filas del DataFrame
df.head()
```

### **3. Creación de etiquetas descriptivas:**

```python
# Normalización de las variables numéricas
df["status_label"] = df["status"].map({
    0: "Normal",
    1: "Incident"
})
# Imprimir la cantidad de registros por cada categoría en la columna "status_label"
print(df["status_label"].value_counts())
```

### **4. Categorización del Uso de CPU:**

```python
# Clasificación de la variable "cpu_utilization" en niveles
df["cpu_level"] = pd.cut(
    df["cpu_utilization"], # Valores de uso de CPU que serán clasificados
    bins=[0, 40, 70, 100], # Definimos los rangos de clasificación para el uso de CPU
    labels=["Low", "Medium", "High"] # Asignamos una etiqueta descriptiva a cada rango
)
# Imprimir la cantidad de registros por cada categoría en la columna "cpu_level"
print(df["cpu_level"].value_counts())
```

### **5. Categorización del uso de Memoria:**

```python
# Clasificación de la variable "memory_level" en niveles
df["memory_level"] = pd.cut(
    df["memory_usage"],  # Valores de uso de memoria que serán clasificados
    bins=[0, 50, 80, np.inf], # Definición de los rangos para cada nivel de uso de memoria
    labels=["Low", "Medium", "High"] # Asignación una etiqueta descriptiva a cada rango
)
# Imprimimos la cantidad de registros por cada categoría en la columna "memory_level"
print(df["memory_level"].value_counts())
```

### **6. Categorización de latencia:**

```python
# Clasificación de la variable "network_latency" en niveles
df["latency_level"] = pd.cut(
    df["network_latency"],  # Valores de latencia de red 
    bins=[0, 50, 100, np.inf], # Definición los Rangos de clasificación
    labels=["Low", "Medium", "High"] # Etiquetas asignadas a cada rango
)
# Imprimimos la cantidad de registros por cada categoría en la columna "latency_level"
print(df["latency_level"].value_counts())
```

### **7. Categorización de temperatura:**

```python
# Clasificación de la variable "temperature" en niveles
df["temperature_level"] = pd.cut(
    df["temperature"], # Valores de ltemperatura 
    bins=[0, 50, 70, np.inf], # Definición los Rangos de clasificación
    labels=["Normal", "Warning", "Critical"] # Etiquetas asignadas a cada rango
)
# Imprimir la cantidad de registros por cada categoría en la columna "temperature_level"
print(df["temperature_level"].value_counts())
```

### **8. Normalización de variables:**

```python
# Vector de nombres de columnas numéricas
columnas_numericas = [
    "cpu_utilization",
    "memory_usage",
    "network_latency",
    "temperature",
    "power_consumption"
]

# Normalización de las variables numéricas
for col in columnas_numericas:
    df[col + "_norm"] = (
        (df[col] - df[col].min())
        / (df[col].max() - df[col].min())
    )
    
    # Imprimir los valores normalizados de cada columna
    print("\n")
    print(f"Valores normalizados de {col}:")  
    print(df[col + "_norm"].head())
```

### **Fase 5. Feature Engineering**
### **9. Creación del indicador de presión del sistema:**

```python
# Categorizar el score de presión del sistema en niveles
df["system_pressure_score"] = (
    df["cpu_utilization_norm"] +
    df["memory_usage_norm"] +
    df["network_latency_norm"] +
    df["temperature_norm"] +
    df["power_consumption_norm"]
) / 5

# Imprimir los valores de presión del sistema
print("Valores del score de presión del sistema:")
print(df["system_pressure_score"].head())
```

### **10. Clasificación del nivel de presión:**

```python
# Clasificación del score de presión del sistema en niveles
df["pressure_level"] = pd.cut(
    df["system_pressure_score"],
    bins=[0, 0.33, 0.66, 1],
    labels=["Low", "Medium", "High"],
    include_lowest=True
)
# Imprimir la cantidad de registros por cada categoría en la columna "pressure_level"
print("Valores del nivel de presión del sistema:")
print(df["pressure_level"].value_counts())
```

### **11. Exportación:**

```python
# Creación de ruta de guardado de DataFrame transformado en un nuevo archivo CSV
ruta_dataframe_tr = "../data/processed"
os.makedirs(ruta_dataframe_tr, exist_ok=True)

# Guardar el DataFrame transformado en un nuevo archivo CSV
df.to_csv(
    "../data/processed/Big_data_dataset_transformado.csv",
    index=False
)
# Imprimir mensaje de confirmación
print("Dataset transformado guardado como 'Big_data_dataset_transformado.csv'")
```

### **12.3. Normalización de Variables Numéricas:**

La normalización es una técnica de transformación de datos que permite convertir variables con diferentes escalas a un rango común entre 0 y 1.

En este proyecto, variables como **la utilización de CPU**, **el uso de memoria**, **la latencia de red**, **la temperatura** y **el consumo energético** presentan magnitudes distintas, lo que puede dificultar su comparación directa y afectar el rendimiento de futuros modelos predictivos.

Para evitar este problema, se aplicó la técnica **Min-Max Scaling**, que transforma cada valor utilizando la siguiente expresión:

Donde:

- \(x\) representa el valor original.
- \(x_{min}\) es el valor mínimo de la variable.
- \(x_{max}\) es el valor máximo de la variable.
- \(x_{norm}\) es el valor normalizado.

Tras la transformación:

- El valor mínimo de cada variable pasa a ser 0.
- El valor máximo pasa a ser 1.
- El resto de los valores quedan distribuidos proporcionalmente dentro de ese rango.

Las variables normalizadas generadas durante esta fase fueron:

- cpu_utilization_norm
- memory_usage_norm
- network_latency_norm
- temperature_norm
- power_consumption_norm

Estas nuevas variables serán utilizadas posteriormente para la construcción del indicador compuesto **system_pressure_score**, así como en posibles modelos de **Machine Learning** donde resulta conveniente trabajar con variables en escalas homogéneas.

La normalización no modifica la información original del dataset, sino que facilita la comparación entre métricas heterogéneas y permite combinar distintas variables en indicadores agregados sin que aquellas con valores numéricamente más altos tengan una influencia desproporcionada sobre los resultados.
