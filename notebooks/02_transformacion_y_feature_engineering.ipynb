### **Fase 4. Transformación de Datos**
### **1. Importación de librerias:**

```python
# Importación de librerias
import pandas as pd # Librería para trabajar con datos en forma de tablas
import numpy as np # Librería para operaciones numéricas
import os # Permite trabajar con rutas y carpetas del sistema
```

### **2. Carga del DataSet:**

```python
# Carga del dataset
df = pd.read_csv("../data/raw/Big_data_dataset.csv")

# Mostrar las primeras filas del DataFrame
df.head()
```

### **3. Creación de etiquetas descriptivas:**

```python
# Normalización de las variables numéricas
df["status_label"] = df["status"].map({
    0: "Normal",
    1: "Incident"
})
# Imprimir la cantidad de registros por cada categoría en la columna "status_label"
print(df["status_label"].value_counts())
```

##### Se creó la variable `status_label` para reemplazar los valores numéricos de la variable objetivo por etiquetas más descriptivas (`Normal` e `Incident`), facilitando la interpretación de los resultados.

##### La distribución obtenida muestra 9.903 registros en estado normal y 97 registros clasificados como incidentes, evidenciando un fuerte desbalance entre ambas clases.

##### Este comportamiento es habitual en entornos de monitorización IT, donde los eventos de fallo son poco frecuentes en comparación con el funcionamiento normal del sistema.

### **4. Categorización del Uso de CPU:**

```python
# Clasificación de la variable "cpu_utilization" en niveles
df["cpu_level"] = pd.cut(
    df["cpu_utilization"], # Valores de uso de CPU que serán clasificados
    bins=[0, 40, 70, 100], # Definimos los rangos de clasificación para el uso de CPU
    labels=["Low", "Medium", "High"] # Asignamos una etiqueta descriptiva a cada rango
)
# Imprimir la cantidad de registros por cada categoría en la columna "cpu_level"
print(df["cpu_level"].value_counts())
```

##### Se categorizó la utilización de CPU en tres niveles (`Low`, `Medium` y `High`) con el objetivo de facilitar el análisis e interpretación del comportamiento del sistema.

##### La distribución obtenida muestra 3.738 registros con uso bajo, 3.212 con uso medio y 3.050 con uso alto, reflejando una representación equilibrada de los distintos niveles de carga del procesador.

##### Estos resultados permiten analizar posteriormente si determinados niveles de utilización de CPU presentan una mayor relación con la aparición de incidentes o estados anómalos del sistema.

### **5. Categorización del uso de Memoria:**

```python
# Clasificación de la variable "memory_level" en niveles
df["memory_level"] = pd.cut(
    df["memory_usage"],  # Valores de uso de memoria que serán clasificados
    bins=[0, 50, 80, np.inf], # Definición de los rangos para cada nivel de uso de memoria
    labels=["Low", "Medium", "High"] # Asignación una etiqueta descriptiva a cada rango
)
# Imprimimos la cantidad de registros por cada categoría en la columna "memory_level"
print(df["memory_level"].value_counts())
```

##### Se categorizó el uso de memoria en tres niveles (`Low`, `Medium` y `High`) para facilitar la interpretación de los distintos estados de utilización de este recurso dentro de la infraestructura IT.

##### La distribución obtenida muestra 4.385 registros con uso bajo, 3.347 con uso medio y 2.268 con uso alto, indicando una mayor concentración de observaciones en niveles reducidos de consumo de memoria.

##### Esta clasificación permite evaluar posteriormente si los niveles elevados de utilización de memoria presentan alguna relación con la aparición de incidentes o situaciones de degradación del sistema.

### **6. Categorización de latencia:**

```python
# Clasificación de la variable "network_latency" en niveles
df["latency_level"] = pd.cut(
    df["network_latency"],  # Valores de latencia de red 
    bins=[0, 50, 100, np.inf], # Definición los Rangos de clasificación
    labels=["Low", "Medium", "High"] # Etiquetas asignadas a cada rango
)
# Imprimimos la cantidad de registros por cada categoría en la columna "latency_level"
print(df["latency_level"].value_counts())
```

##### Se categorizó la latencia de red en tres niveles (`Low`, `Medium` y `High`) con el fin de simplificar el análisis del rendimiento de las comunicaciones dentro de la infraestructura IT.

##### La distribución obtenida muestra 5.027 registros con latencia alta, 2.491 con latencia baja y 2.482 con latencia media, observándose una mayor concentración de observaciones en niveles elevados de latencia.

##### Esta clasificación facilita la identificación de posibles patrones entre el comportamiento de la red y la aparición de incidentes, permitiendo evaluar el impacto de la latencia en la estabilidad del sistema.

### **7. Categorización de temperatura:**

```python
# Clasificación de la variable "temperature" en niveles
df["temperature_level"] = pd.cut(
    df["temperature"], # Valores de ltemperatura 
    bins=[0, 50, 70, np.inf], # Definición los Rangos de clasificación
    labels=["Normal", "Warning", "Critical"] # Etiquetas asignadas a cada rango
)
# Imprimir la cantidad de registros por cada categoría en la columna "temperature_level"
print(df["temperature_level"].value_counts())
```

##### Se categorizó la temperatura del sistema en tres niveles (`Normal`, `Warning` y `Critical`) para facilitar la identificación de posibles situaciones de riesgo relacionadas con el sobrecalentamiento de la infraestructura IT.

##### La distribución obtenida muestra 3.104 registros en estado normal, 3.101 en nivel de advertencia y 3.795 en estado crítico, observándose una ligera predominancia de temperaturas elevadas dentro del conjunto de datos.

##### Esta clasificación permite analizar posteriormente si los niveles de temperatura más altos presentan una mayor asociación con la aparición de incidentes, degradación del rendimiento o posibles fallos operativos del sistema.

### **8. Normalización de variables:**

```python
# Vector de nombres de columnas numéricas
columnas_numericas = [
    "cpu_utilization",
    "memory_usage",
    "network_latency",
    "temperature",
    "power_consumption"
]

# Normalización de las variables numéricas
for col in columnas_numericas:
    df[col + "_norm"] = (
        (df[col] - df[col].min())
        / (df[col].max() - df[col].min())
    )
    
    # Imprimir los valores normalizados de cada columna
    print("\n")
    print(f"Valores normalizados de {col}:")  
    print(df[col + "_norm"].head())
```

##### Se aplicó un proceso de normalización a variables clave del sistema, como la utilización de CPU, uso de memoria, latencia de red, temperatura y consumo energético, transformando sus valores a una escala común entre 0 y 1.

##### Los resultados muestran que todas las variables fueron correctamente escaladas, permitiendo comparar métricas originalmente expresadas en unidades y rangos diferentes sin que ninguna tenga un peso desproporcionado en los análisis posteriores.

##### Esta transformación resulta especialmente útil para la creación de indicadores compuestos y para el entrenamiento de modelos de Machine Learning, ya que mejora la comparabilidad entre variables y favorece la estabilidad de los algoritmos predictivos.

### **Fase 5. Feature Engineering**
### **9. Creación del indicador de presión del sistema:**

```python
# Categorizar el score de presión del sistema en niveles
df["system_pressure_score"] = (
    df["cpu_utilization_norm"] +
    df["memory_usage_norm"] +
    df["network_latency_norm"] +
    df["temperature_norm"] +
    df["power_consumption_norm"]
) / 5

# Imprimir los valores de presión del sistema
print("Valores del score de presión del sistema:")
print(df["system_pressure_score"].head())
```

##### Se creó el indicador `system_pressure_score`, calculado a partir de la combinación de variables normalizadas como utilización de CPU, uso de memoria, latencia de red, temperatura y consumo energético.

##### Los valores obtenidos oscilan entre 0 y 1, donde puntuaciones más elevadas representan una mayor presión operativa sobre la infraestructura IT y puntuaciones más bajas reflejan condiciones de funcionamiento más estables.

##### Este indicador permite resumir múltiples métricas técnicas en una única variable, facilitando el análisis global del estado del sistema y la identificación de posibles situaciones de sobrecarga o riesgo operativo.

### **10. Clasificación del nivel de presión:**

```python
# Clasificación del score de presión del sistema en niveles
df["pressure_level"] = pd.cut(
    df["system_pressure_score"],
    bins=[0, 0.33, 0.66, 1],
    labels=["Low", "Medium", "High"],
    include_lowest=True
)
# Imprimir la cantidad de registros por cada categoría en la columna "pressure_level"
print("Valores del nivel de presión del sistema:")
print(df["pressure_level"].value_counts())
```

##### Se clasificó el indicador `system_pressure_score` en tres niveles de presión (`Low`, `Medium` y `High`) con el objetivo de facilitar la interpretación del estado operativo general de la infraestructura IT.

##### La distribución obtenida muestra 965 registros con presión baja, 7.934 con presión media y 1.101 con presión alta, evidenciando que la mayoría de los sistemas operan bajo condiciones de carga moderada.

##### Esta categorización permite identificar de forma más sencilla situaciones potenciales de sobrecarga y analizar posteriormente si los niveles elevados de presión presentan una mayor relación con la aparición de incidentes o degradaciones del rendimiento.

### **12.3. Normalización de Variables Numéricas:**

La normalización es una técnica de transformación de datos que permite convertir variables con diferentes escalas a un rango común entre 0 y 1.

En este proyecto, variables como **la utilización de CPU**, **el uso de memoria**, **la latencia de red**, **la temperatura** y **el consumo energético** presentan magnitudes distintas, lo que puede dificultar su comparación directa y afectar el rendimiento de futuros modelos predictivos.

Para evitar este problema, se aplicó la técnica **Min-Max Scaling**, que transforma cada valor utilizando la siguiente expresión:

Donde:

- \(x\) representa el valor original.
- \(x_{min}\) es el valor mínimo de la variable.
- \(x_{max}\) es el valor máximo de la variable.
- \(x_{norm}\) es el valor normalizado.

Tras la transformación:

- El valor mínimo de cada variable pasa a ser 0.
- El valor máximo pasa a ser 1.
- El resto de los valores quedan distribuidos proporcionalmente dentro de ese rango.

Las variables normalizadas generadas durante esta fase fueron:

- cpu_utilization_norm
- memory_usage_norm
- network_latency_norm
- temperature_norm
- power_consumption_norm

Estas nuevas variables serán utilizadas posteriormente para la construcción del indicador compuesto **system_pressure_score**, así como en posibles modelos de **Machine Learning** donde resulta conveniente trabajar con variables en escalas homogéneas.

La normalización no modifica la información original del dataset, sino que facilita la comparación entre métricas heterogéneas y permite combinar distintas variables en indicadores agregados sin que aquellas con valores numéricamente más altos tengan una influencia desproporcionada sobre los resultados.
